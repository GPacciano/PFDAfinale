```python
import pandas as pd
from matplotlib import pyplot as plt

#This data is for production of bottles and cans
beerbottles_cans= pd.read_csv("https://raw.githubusercontent.com/GPacciano/PFDAfinale/main/BottlesCanscsv.csv")

#Subset of NY Bottles and Cans Production
NYcanbot= beerbottles_cans[beerbottles_cans['state'] == "NY"]
print("NY Bottle and Can Production:",)
print(NYcanbot)
canbot2008= NYcanbot[NYcanbot['year'] == 2008]
canbot2013= NYcanbot[NYcanbot['year'] == 2013]
canbot2019= NYcanbot[NYcanbot['year'] == 2019]
canbot2008pro= canbot2008['barrels']
canbot2013pro= canbot2013['barrels']
canbot2019pro= canbot2019['barrels']

#Mean, Median, of NY production of Cans and Bottles
canbotmean= beerbottles_cans['barrels'].mean()
canbotmedian= beerbottles_cans['barrels'].median()
print("This is the mean produced in cans and bottles from 2008 to 2019: ", canbotmean)
print("This is the median produced in cans and bottles from 2008 to 2019: ", canbotmedian)


#Scatter plot of Can and Bottle Production
canyear=NYcanbot['year']
canbarrels=NYcanbot['barrels']
plt.scatter(canyear, canbarrels, label='Scatter Plot', color='orange', marker='o')
plt.xlabel('Year')
plt.ylabel('Barrels Produced')
plt.title('NY Beer Production in Cans and Bottles')
plt.figure(figsize=(50, 30))
plt.show()

#This data is for kegs and barrels
beerkegs_barrels= pd.read_csv("https://raw.githubusercontent.com/GPacciano/PFDAfinale/main/KegsBarrelscsv.csv")

#Subset of NY Kegs and Barrels Production
NYkegbar= beerkegs_barrels[beerkegs_barrels['state'] == "NY"]
print("NY Keg and Barrel Production:")
print(NYkegbar)
kegbar2008= NYkegbar[NYkegbar['year'] == 2008]
kegbar2013= NYkegbar[NYkegbar['year'] == 2013]
kegbar2019= NYkegbar[NYkegbar['year'] == 2019]
kegbar2008pro= kegbar2008['barrels']
kegbar2013pro= kegbar2013['barrels']
kegbar2019pro= kegbar2019['barrels']

#Mean, Median, of NY production of Cans and Bottles
kegbarrelmean= NYkegbar['barrels'].mean()
kegbarrelmedian= NYkegbar['barrels'].median()
print("This is the mean produced in kegs and barrels from 2008 to 2019: ", kegbarrelmean)
print("This is the median produced in kegs and barrels from 2008 to 2019: ", kegbarrelmedian)


#Scatter plot of keg and barrel production
nykegbaryear=NYkegbar['year']
nykegbarbarrels=NYkegbar['barrels']
plt.scatter(nykegbaryear, nykegbarbarrels, label='Scatter Plot', color='blue', marker='o')
plt.xlabel('Year')
plt.ylabel('Barrels Produced')
plt.title('NY Beer Production in Kegs and Barrels')
plt.figure(figsize=(50, 30))
plt.show()

#This data is for on premises production
beer_premises= pd.read_csv("https://raw.githubusercontent.com/GPacciano/PFDAfinale/main/OnPremisescsv.csv")

#Subset of NY beer produced on premises
NYprem= beer_premises[beer_premises['state'] == "NY"]
print("NY on Premises Production:")
print(NYprem)
nyprem2008= NYprem[NYprem['year'] == 2008]
nyprem2013= NYprem[NYprem['year'] == 2013]
nyprem2019= NYprem[NYprem['year'] == 2019]
nyprem2008pro= nyprem2008['barrels']
nyprem2013pro= nyprem2013['barrels']
nyprem2019pro= nyprem2019['barrels']

#Mean, Median, of NY production of Cans and Bottles
nyprem_mean= NYprem['barrels'].mean()
nyprem_median= NYprem['barrels'].median()
print("This is the mean produced on the premises from 2008 to 2019: ", nyprem_mean)
print("This is the median produced on the premises from 2008 to 2019: ", nyprem_median)

#Scatter plot of NY beer produced on the premises
nypremisesyear=NYprem['year']
nypremisesbarrels=NYprem['barrels']
plt.scatter(nypremisesyear, nypremisesbarrels, label='Scatter Plot', color='green', marker='o')
plt.xlabel('Year')
plt.ylabel('Barrels Produced')
plt.title('NY Beer Production on Premises')
plt.figure(figsize=(50, 30))
plt.show()

#This is a bar graph of all 3 means of production
fig, axes= plt.subplots (1,3, figsize= (18,6))
axes[0].bar(canyear,canbarrels, color='orange', alpha=0.7)
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Barrels')
axes[0].set_title('NY Beer Production in Cans and Bottles')

axes[1].bar(nykegbaryear,nykegbarbarrels, color='blue', alpha=0.7)
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Barrels')
axes[1].set_title('NY Beer Production in Kegs and Barrels')

axes[2].bar(nypremisesyear, nypremisesbarrels, color='green', alpha=0.7)
axes[2].set_xlabel('Year')
axes[2].set_ylabel('Barrels')
axes[2].set_title('NY Beer Production on Premises')

plt.tight_layout()
plt.show()

#Pie chart to show share of production for 2008
labels= ['Cans & Bottles', 'Kegs & Barrels', 'On Premises']
piecolors = ['orange', 'blue', 'green']
plt.figure(figsize=(6, 6))
plt.pie([6997372.81, 749430.78, 12354.42], labels=labels, autopct='%1.1f%%', startangle=140, shadow=True, colors=piecolors)
plt.title('2008 Production')
plt.show()

#Pie chart to show share of production for 2013
labels= ['Cans & Bottles', 'Kegs & Barrels', 'On Premises']
piecolors = ['orange', 'blue', 'green']
plt.figure(figsize=(6, 6))
plt.pie([7166945.25, 674239.50, 24800.93], labels=labels, autopct='%1.1f%%', startangle=140, shadow=True, colors=piecolors)
plt.title('2013 Production')
plt.show()

#Pie chart to show share of production for 2019
labels= ['Cans & Bottles', 'Kegs & Barrels', 'On Premises']
piecolors = ['orange', 'blue', 'green']
plt.figure(figsize=(6, 6))
plt.pie([7720584.90, 450143.63, 150616.79], labels=labels, autopct='%1.1f%%', startangle=140, shadow=True, colors=piecolors)
plt.title('2019 Production')
plt.show()
```
